```python
import pandas as pd
```

```python
data = pd.read_csv('FullDF_duplicated.csv',encoding='utf-8-sig',index_col=0)
data
```

```python
data = data.drop('리뷰',axis=1)
```

```python
data = data.replace(to_replace =50, value =5)
data = data.replace(to_replace =40, value =4) 
data = data.replace(to_replace =30, value =3) 
data = data.replace(to_replace =20, value =2) 
data = data.replace(to_replace =10, value =1) 
```

```python
data
```

```python
df_table = data.set_index(["작성자",'0']).unstack()
df_table
```

```python
df_table.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ibc = cosine_similarity(df_table)
ibc = pd.DataFrame(data=ibc, index=df_table.index, columns=df_table.index)
ibc
```

```python
import surprise
from surprise import Reader
from surprise import Dataset
from surprise.model_selection import train_test_split
from surprise import accuracy
reader = Reader(rating_scale=(1,5))
df = Dataset.load_from_df(data[['작성자','0','평점']],reader)
```

```python
sim_options = {'name':'cosine','user_based':'True'}
trainset, testset = train_test_split(df,test_size=0.25)
```

```python
ubcf = surprise.KNNBasic(sim_options=sim_options)
```

```python
from surprise.model_selection import cross_validate
cross_validate(ubcf, df, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
sim_options = {'name':'cosine','user_based':'False'}
ibcf = surprise.KNNBasic(sim_options=sim_options)
```

```python
from surprise.model_selection import cross_validate
cross_validate(ibcf, df, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
